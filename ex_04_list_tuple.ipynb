```python
### 리스트 선언하기

a = [] # 빈 리스트
b = [1,2,3] # 숫자 데이터를 담은 리스트
c = ['my', 'name', 'is', 'jy'] # 문자열 데이터를 담은 리스트
d = [ 1, 2, 'my', 'name' ] # 숫자열, 문자열 데이터를 담은 리스트
e = [ 1, 2, [ 'my', 'name' ]] # 이중 리스트 가능하다.

print(a)
print(b)
print(c)
print(d)
print(e)
print(type(e))
```

#### 리스트 데이터 접근
    - 인덱스 번호 활용
    - 문자열 인덱싱, 슬라이싱 활용과 동일
    - 인덱싱 : 무엇인가를 '가리킴'
    - 슬라이싱 : 무엇인가를 '잘라냄' (범위)
    - list[index], list[:]

```python
list1 = [ 2, 5, 7, 9, 10 ]

# list 에서 숫자 2 에 접근
print(list1[0])

# list 에서 숫자 9 에 접근
print(list1[3])
```

```python

list2 = [ 1, 2, 3, [ 'a', 'b', 'c ']]

# 변수 temp 에 ['a', 'b', 'c' ] 를 저장하고 출력하시오
temp = list2[3]
print(temp)
#  list2 에서 문자 'b'만 뽑아서 출력하시오.
print(list2[3][1])
print(temp[1])
```

```python
# 슬라이싱
# list[시작:끝+1]

list3 = [ '지수', '문규', '다인', '수양', '성문']

print(list3[0:2])

# 슬라이싱 결과는 여러개에 대해 리스트로 묶어서 반환한다.

# '다인', '성문' 슬라이싱 해보기

print(list3[2:5:2])

# 맨 뒤에부터 앞에까지 차례대로 출력하기
list3[::-1]
```

```python
# 리스트 가변적이다.
# 리스트 더하기

list4 = [ '인우', '지웅', '강련', '서준희']

print(list3+list4)

# 리스트 곱하기

list5 = [ 1, 2, 3 ]

print(list5 * 3) # 수학 개념의 연산이 아니다. 자료의 크기를 그만큼 늘린다.
```

```python
# 리스트 맨 뒤쪽 값 추가
# 리스트.append(값)

list5.append(4) # list5 의 마지막에 값을 추가한다.

print(list5)
```

```python
fuits = [ '사과', '포도', ['수박', '멜론'], '오렌지']

apple = fuits[0]
print(apple)
melon = fuits[2][1]
print(melon)
orange = fuits[3]
print(orange)

choice_list= []
choice_list.append(apple)
choice_list.append(melon)
choice_list.append(orange)
print(choice_list)
```

```python
# 리스트 길이 확인

len(choice_list) # 요소 1개 --> [] 리스트가 요소값 1개
```

```python
# 리스트 값 수정
array = [ 1, 2, 3, 4, 5]

# Q1. 5 > 7 로 바꾸기
# 변경될 위치 접근
# 새로운 값을 대입
array[-1] = 7
print(array[-1])
array[1] = ['a', 'b', 'c']
print(array)

# list 안의 요소 삭제하기
# del 키워드 삭제
# del list[index]
del array[3]
print(array)

# list 자체 함수
# list.remove(value)
# value 값이 존재하지 않으면 error 발생한다.
# Remove first occurrence of value.
# Raises ValueError if the value is not present.
array.remove(3)
print(array)

# list.pop()
# Remove and return item at index (default last).
# Raises IndexError if list is empty or index is out of range.
print(array.pop())
print(array)
```

```python
# 튜플 선언하기
a = () # 빈 튜플
b = ( 1, 2, 3) # 숫자를 담아서 선언
c = ( 'my', 'name') # 문자를 담아서 선언
d = ( 1, 2, 'my') # 다양한 자료형을 담아서 선언
e = ( 1, 2, ( 3, 4, 'name'), 5) # 이중 튜플 선언 가능
print(a, b, c, d, e)
```

```python
# 튜플 수정이 불가능
# 튜플 조회, 접근만 가능
# 인덱싱, 슬라싱
tp1 = ('건욱', '한준희', '근영', '기원', '인우')

# tp1[0] = '장건'
# TypeError: 'tuple' object does not support item assignment

# 준희:기원
print(tp1[1:-1])
print(tp1[1:4])
```

```python
# list.index(value) --> return index
lst = [ '딸기', '수박', '바나나', '체리', '포도']
search = input('검색할 문자를 입력하세요 >> ')

lstIndex = -1
try:
    lstIndex = lst.index(search) # 없을 경우 exception 발생
except:
    print('List does not contain value')
    
if (lstIndex >= 0):
    print(f'{search} 는 리스트에 {lstIndex} 번째 인덱스에 들어있습니다.')
else:    
    print(f'{search} 는 리스트에 들어있지 않습니다.')
    
if (search in lst):
    print(f'{search} 는 리스트에 {lstIndex} 번째 인덱스에 들어있습니다.')
else:
    print(f'{search} 는 리스트에 들어있지 않습니다.')
    
print('들어있다' if search in lst else '없다')
    
    
```

```python
# 문자열대상.count(value)
s = 'Hi, My name is Mihee'

inputValue = input("검색할 문자를 입력하세요 >> ")

valueCount = s.count(inputValue)
if (valueCount > 0 or inputValue in s):
    print(f'\'{inputValue}\'는 문자열에 {valueCount} 번 포함되어 있습니다.')
else:
    print(f'\'{inputValue}\'는 문자열에 포함되어있지 않습니다.')
```

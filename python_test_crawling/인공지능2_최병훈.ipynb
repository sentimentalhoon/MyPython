# 문제3
자동제어를 통해 G마켓 홈페이지에 접속하고, 조건에 맞는 수집프로그램을 만드시오.

```python
driver = wb.Chrome()
driver.implicitly_wait(10)

driver.get('http://corners.gmarket.co.kr/Bestsellers')
time.sleep(3)
itemNo = driver.find_elements(By.CSS_SELECTOR, '#gBestWrap > div.best-list > ul > li > p')
itemName = driver.find_elements(By.CSS_SELECTOR, '#gBestWrap > div.best-list > ul > li > a')
itemPrice  = driver.find_elements(By.CSS_SELECTOR, '#gBestWrap > div.best-list > ul > li > div.item_price > div.s-price > strong > span')

itemsNo = []
itemsName = []
itemsPrice = []
for no in tqdm(itemNo):
    intNo = int(no.text)
    if intNo <= 20:
        itemsNo.append(itemNo[intNo - 1].text)
        itemsName.append(itemName[intNo - 1].text)
        itemsPrice.append(itemPrice[intNo - 1].text)
    elif intNo > 20:
        break        

gmarketBest20DF = pd.DataFrame({ '순위' : itemsNo, '상품명' : itemsName, '가격' : itemsPrice })
gmarketBest20DF.set_index('순위', inplace=True)  
display(gmarketBest20DF)

driver.quit()
```

# 문제4
멜론차트에서 1~100위까지 정보를 수집하는 프로그램을 작성하시오.

```python
############ beautifulsoup 사용
url = 'https://www.melon.com/chart/index.htm'
headers = {
    'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/115.0.0.0 Safari/537.36'
}
# 페이지 호출
res = requests.get(url, headers = headers)
soup = bs(res.content.decode('utf-8', 'replace'), 'lxml')
topTitleList = []
topSingerList = []
topRankList = []

title = soup.select('div.ellipsis.rank01 > span > a')
singer = soup.select('div.ellipsis.rank02 > span.checkEllipsis')
    
for i in tqdm(range(len(singer))):
    topSingerList.append(singer[i].text)
    topTitleList.append(title[i].text)
    topRankList.append(i+1)
    
melon = pd.DataFrame({'순위' : topRankList, '곡명' : topTitleList, '가수' : topSingerList})
melon.set_index('순위', inplace=True)  
display(melon)

now = datetime.now()
file_date_name = f'{now.date()}_{now.hour}_{now.minute}_{now.second}'
melon.to_csv(f'melon_top_100_{file_date_name}.csv', encoding='utf-8')
```

```python
############## selenium 사용.... 으흠... 왜 가수가 안 읽어와질까나.........

driver = wb.Chrome()
driver.implicitly_wait(10)
try:
        driver.get('https://www.melon.com/chart/index.htm')
        time.sleep(3)

        titles = driver.find_elements(By.CSS_SELECTOR, 'div.ellipsis.rank01 > span > a')
        singers = driver.find_elements(By.CSS_SELECTOR, 'div.ellipsis.rank02 > span > a.checkEllipsis')

        rank_List = []
        title_List = []
        singer_List = []
        for i in tqdm(range(len(title))):
                rank_List.append(i+1)
                title_List.append(titles[i].text)
                singer_List.append(singers[i].text)
                
        melonTop100 = pd.DataFrame({ '순위' : rank_List, '곡명' : title_List, '가수' : singer_List })
        melonTop100.set_index('순위', inplace=True)  
        display(melonTop100)

        now = datetime.now()

        file_date_name = f'{now.date()}_{now.hour}_{now.minute}_{now.second}'
        melonTop100.to_csv(f'melon_top_100_{file_date_name}.csv', encoding='utf-8')        
except:
        print('error')
finally:
        driver.quit()                
```

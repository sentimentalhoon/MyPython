```python
dic1 = { 'name' : 'jy', 'age' : 20, 'phone' : '010-1111-1234' }

# dictionary 값 추가

dic1['birth'] = '02/11'

# 데이터 접근하기
# 딕셔너리 값 가져오기
# dictionary[key] or dictionary.get(key)

# 생일 데이터 접근
dic1['birth']
dic1.get('birth')

# 이름 접근하기
dic1['name']
dic1.get('name')

# 모든 key 확인하기
dic1.keys()
# 모든 value 확인
dic1.values()

# dic2 -> 우리반 친구들의 정보를 담은 딕셔너리
class_info = { 'name' : [ '현석', '수양', '나행'],
              'age' : [ 28, 20, 25 ],
              'blood' : [ 'B', 'A', 'O' ]
              }

# dictionary 는 for 문으로 이용 가능
for key in class_info.keys():
    print(key)
for value in class_info.values():
    print(value)
for key, value in class_info.items():
    print(f'{key} : {value[0]}')
      

    
```

```python
score_dict = {'이름':['손지영','김은영','김운비','채수민','최성우'],
        'Python':[90,100,85,90,80],
        'Java':[85,80,100,95,85],
        'html/css':[75,70,90,80,90]}

# 1. python 과목 평균 구하기
pyTotal = score_dict['Python']
print(sum(pyTotal)/len(pyTotal))

score_dic_key = list(score_dict.keys())
print(score_dic_key[1:])

for i in score_dic_key[1:]:
    avg = sum(score_dict.get(i)) / len(score_dict.get(i))
    print(f'{i} 과목의 평균 : {avg}')    
    
for data in score_dict:
    # data -> key return
    # value return
    if (data == '이름'):
        continue
    
    classPlus = sum(score_dict.get(data))
        
    print(f'{data} 과목의 평균 : {classPlus/len(score_dict.get(data))}')
```

```python
inputName = input("이름을 정확하게 기입하여 주십시요 >> ")

indexNum = -1
b = False
for name in score_dict['이름']:
        indexNum += 1
        if inputName == name:
                b = True
                break
        
if b:
        userAllScore = 0
        userAvg = 0

        for i in score_dic_key[1:]:        
                userAllScore += score_dict.get(i)[indexNum]

        userAvg = round(userAllScore / (len(score_dic_key) -1), 2)
        print(f'{inputName} 님의 전 과목 성적 평균 : {userAvg}')  
else:                      
        print('검색 결과가 없습니다.')
                        
```

```python
inputName = input("이름을 정확하게 기입하여 주십시요 >> ")

try:
        indexNum = score_dict['이름'].index(inputName)
        userAllScore = 0
        userAvg = 0

        for i in score_dic_key[1:]:        
                userAllScore += score_dict.get(i)[indexNum]

        userAvg = round(userAllScore / (len(score_dic_key) -1), 2)
        print(f'{inputName} 님의 전 과목 성적 평균 : {userAvg}')  
except ValueError:
        print('검색 결과가 없습니다.')
```

```python
s_sum = 0
name = input('이름을 정확하게 입력>> ')
# 유무를 체크하는 변수 
check = False

for i in range(0,len(score_dict['이름'])):
    if (score_dict['이름'][i] == name):
        check = True # 정보가 있었음 , 이름이 찾아졌음
        s_sum += score_dict['Python'][i]
        s_sum += score_dict['Java'][i]
        s_sum += score_dict['html/css'][i]
        print(f'{name}님의 전 과목 성적 평균: {round(s_sum/3,2)}')

if not check: 
    print('검색 결과가 없습니다.')
```

#### 반복문 : 반복하여 수행하는 명령문
    - while, for

```python
number = 1

while number <= 3:
    print(number)
    number += 1
```

```python
# while 문 사용해서 "파이썬이 쉬워요." 문장을 13번 출력하기.

s = '파이썬이 쉬워요.'

number = 13
while number > 0:
    print(number, s)
    number -= 1
    
number = 13    
while True:
    print(number, s)
    number -= 1
    if (number <= 0):
        break
```

```python
while True:
    num1 = int(input("첫 번째 정수 입력 >> "))
    num2 = int(input("두 번째 정수 입력 >> "))
    plusNum = num1 + num2
    if (plusNum == 0):
        print('프로그램이 종료되었습니다.')
        break    
    print(f'두 정수의 합 : {plusNum}')
```

```python
import cv2
cap = cv2.VideoCapture(0) # 0번째 카메라 장치( Device ID , 실행창 devmgmt.msc>카메라>속성)

# 코덱 정의
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps =  30.0  # 초당 프레임 이미지 수

# 저장 파일명, 코덱, fps, 크기(가로, 세로)
out = cv2.VideoWriter('out.avi', fourcc, fps, (width, height))

# 녹화 상태 여부
record = False

if not cap.isOpened(): # 카메라가 잘 열리지 않은 경우
    exit() # 프로그램 종료
    
while True:
    ret, frame = cap.read()
    if not ret:
        print("비디오 읽기 실패 또는 비디오 모두 읽음")
        break
    out.write(frame)
    cv2.imshow('camera', frame)
    
    if record :
        out.write(frame)
        
    if cv2.waitKey(30) == ord('q'): # 사용자가 알파벳 q를 눌렀을 때 카메라 끄기
        print('카메라 끄기')
        out.release()
        cap.release()
        cv2.destroyAllWindows()
        break
    elif cv2.waitKey(30) == ord('w'):
        print('녹화 시작!')
        record = True
```

```python
import cv2
cap = cv2.VideoCapture(0) # 0  ( Device ID ,  devmgmt.msc>>)

#  
fourcc = cv2.VideoWriter_fourcc(*'DIVX')

width = round(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = round(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps =  30.0  #    

#  , , fps, (, )
out = cv2.VideoWriter('out.avi', fourcc, fps, (width, height))

#   
record = False

if not cap.isOpened(): #     
    exit() #  
    
while True:
    ret, frame = cap.read()
    if not ret:
        print("      ")
        break
    out.write(frame)
    cv2.imshow('camera', frame)
    
    if record :
        out.write(frame)
        
    if cv2.waitKey(30) == ord('q'): #   q    
        print(' ')
        out.release()
        cap.release()
        cv2.destroyAllWindows()
        break
    elif cv2.waitKey(30) == ord('w'):
        print(' !')
        record = True
```

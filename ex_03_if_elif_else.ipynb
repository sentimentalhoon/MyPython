### 조건문 : 상황에 따라 실행 흐름을 다르게 하는 문법

```python
# 조건의 종류 : if, elif, else
# if : 조건식이 True 일 경우 실행문장 실행
# 영역구분은 들여쓰기로 구분한다.
# if 조건식 :
# elif 조건식 :
# else:
a = 1
b = 3
if a > b:
    print('a 가 b 보다 크다')
elif a < b:
    print('a 가 b 보다 작다')
else:
    print('a 와 b 는 같다')
```

```python
# 들여쓰기의 중요성
# out : 조건문 안 실행문장
# out : if 문 실행문장
if True:
    print('조건문 안 실행문장')
print('if 문 실행문장')
```

```python
# out : if 문 실행문장
if False:
    print('조건문 안 실행문장')
print('if 문 실행문장')
```

```python
menuAndPrice = [ 0, ['콜라', 600], ['사이다', 800], ['물', 1000]]
print('--------------- 자 판 기 ---------------')
print(f' 1. {menuAndPrice[1][0]} : {menuAndPrice[1][1]}원\t2. {menuAndPrice[2][0]} : {menuAndPrice[2][1]}원\t3.{menuAndPrice[3][0]} : {menuAndPrice[3][1]}원')

inputMoney = int(input('금액넣기 >> '))
seletMemu = int(input('메뉴선택 >> '))

print(f'현재 보유금액 : {inputMoney}')
print(f'선택한 메뉴 번호 : {seletMemu}')

if (menuAndPrice[seletMemu][1] <= inputMoney):
    inputMoney -= menuAndPrice[seletMemu][1]
    a1000 = int(inputMoney / 1000)
    a500 = int(inputMoney % 1000 / 500)
    a100 = int(inputMoney % 1000 % 500 / 100)
    print(f'거스름돈 >> 1000원 {a1000} 개, 500원 {a500} 개, 100원 {a100} 개')
else:    
    print('금액이 부족합니다.')
```

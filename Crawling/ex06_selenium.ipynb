```python
from selenium import webdriver as wb
# 컴퓨터용 키보드 라이브러리
from selenium.webdriver.common.keys import Keys

# 선택자를 구분지어주는 라이브러리
from selenium.webdriver.common.by import By

import time
```

```python
driver = wb.Chrome()

driver.get('http://www.naver.com')

# https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%A5%98%ED%98%84%EC%A7%84&sort=0&photo=0&field=0&pd=0&ds=&de=&cluster_rank=30&mynews=0&office_type=0&office_section_code=0&news_office_checked=&nso=so:r,p:all,a:all&start=1
searchStr = ''
search = driver.find_element(By.ID, 'query')

search.send_keys('류현진')

search.send_keys(Keys.ENTER)

driver.find_element(By.CSS_SELECTOR, '#lnb > div.lnb_group > div > ul > li:nth-child(2) > a').click()
```

```python
newTitle = []
for i in range(15):
    driver.get(f'https://search.naver.com/search.naver?where=news&sm=tab_pge&query=%EB%A5%98%ED%98%84%EC%A7%84&ds=2023.08.20.10.04&de=2023.08.21.10.04&start={i * 10 + 1}')
    #sp_nws128 > div.news_wrap.api_ani_send > div > a
    news_tit = driver.find_elements(By.CSS_SELECTOR, 'a.news_tit')
    if news_tit is None:
        break
    
    for new in news_tit:
        news = []
        news.append(new.get_attribute('href'))
        news.append(new.text)
        
        newTitle.append(news)
        
    time.sleep(3)
```

```python
for i in newTitle:
    print(i)
```

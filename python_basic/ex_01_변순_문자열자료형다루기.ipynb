### 문자열에 접근하는 방법
 - 문자열 --> 데이터
 - 인덱싱, 슬라이싱

```python
s = "my name is jy"
print(s[0]) # 0
print(s[8]) # i
print(s[-2]) # j
print(s[-1]) # j

# Slicing -> 특정 문자열에 접근
# [시작인덱스:끝인덱스+1]
print(s[0:2]) # my
print(s[3:7]) # name
print(s[8:10]) # is
# Step 기능
print(s[3:7:2]) # nm
print(s[8:10:2]) # i
```

```python
# 생략할 수 있는 인덱스 번호
# s 변수에서 맨 첫번째 인덱스부터 name | e 까지 포함해서 슬라이싱
print(s[:7]) # myname # 시작값의 기본값 : 0 (생략가능)
print(s[0:7]) # myname
print(s[-5:]) # is jy
print(s[8:]) #is jy # 끝값의 기본값 : 문장의 끝 (생략가능)
print(s[:]) # my name is jy # 모든 문자열에 접근
print(s[::2]) # m aei y # index 0부터 시작하여 2씩 증가한다
```

#### 문자열 관련 함수 정리

```python
# 문자열의 길이 확인하는 함수
len(now)
now[4:len(now)] # '7월 26일입니다'
```

```python
# 문자열 대소문자 변환 관련 함수
eng_str = 'hi, My name is JY'
# 소문자를 대문자로 바꾸기
eng_str2 = eng_str.upper()
print(eng_str2) # HI, MY NAME IS JY
#대문자를 소문자로 바꾸기
eng_str3 = eng_str.lower()
print(eng_str3) # hi, my name is jy
```

```python
# '찾을 문자열' 이 몇개 있는지 확인하는 함수
eng_str = 'hi, My name is JY'
eng_str.count('m') # 1
eng_str.count('M') # 1
eng_str.lower().count('m') # 2
```

```python
# 특정 문자열 찾아서 인덱스 번호 확인하는 함수
# find()
# index()
eng_str.find('i') # 1 : 제일 먼저 찾아지는 'i' 인덱스 값 출력
eng_str.find('is') # 12 : 제일 먼저 찾아지는 'i' 인덱스 값 출력

# 문자열에 없는 단어 찾아달라고 하기
eng_str.find('z') # -1 제일 먼저 찾아지는 'i' 인덱스 값 출력
# find 는 해당 단어가 없으면 -1 값 반환, 에러가 발생하지 않는다.
eng_str.index('z') # 제일 먼저 찾아지는 'i' 인덱스 값 출력
# index 는 해당 단어가 없으면 에러가 발생한다.
```

```python
# strip() 문자열 양끝의 공백을 제거하는 함수
'        배고프다         '.strip()
```

```python
# 문자열 1을 문자열 2로 변경하는 함수
# replace(문자열1, 문자열2)
eng_str.replace('JY' , 'JH') # 'hi, My name is JY' --> 'hi, My name is JH'
```

```python
# list.index(value) --> return index
lst = [ '', '', '', '', '']
search = input('   >> ')

lstIndex = -1
try:
    lstIndex = lst.index(search) #   exception 
except:
    print('List does not contain value')
    
if (lstIndex >= 0):
    print(f'{search}   {lstIndex}   .')
else:    
    print(f'{search}    .')
    
if (search in lst):
    print(f'{search}   {lstIndex}   .')
else:
    print(f'{search}    .')
    
print('' if search in lst else '')
    
    
```

```python
# .count(value)
s = 'Hi, My name is Mihee'

inputValue = input("   >> ")

valueCount = s.count(inputValue)
if (valueCount > 0 or inputValue in s):
    print(f'\'{inputValue}\'  {valueCount}   .')
else:
    print(f'\'{inputValue}\'   .')
```

```python
# global 키워드

x = 9

print('전역 변수 : ', x)

def temp():
    global x # x 를 전역 변수로 만든다.
    x = 10
    print('전역 변수를 개념으로 사용 : ', x)
    
temp()

print('전역 변수2 : ', x)    
```

```python
def divisor(n):   
    """ 약수 구하기 함수이다. """ 
    for i in range(1, int(n / 2 + 1)):
        if (n % i == 0):
            print(i, end=" ")
    print(n) 
            
divisor(25)            
```

```python
def cal(n1, n2, op):
    """ 
        사칙 연산을 계산하는 함수입니다.
        num1 : int
        num2 : int
        op : string 연산자 ( +, -, *, / )
    """
    result = 0
    if (op == '+'):
        return n1 + n2
    if (op == '-'):
        return n1 - n2
    if (op == '*'):
        return n1 * n2
    if (op == '/'):
        return n1 / n2
    if (op == '%'):
        return n1 % n2
    return 0
    
def calTernary(n1, n2, op):
    """ 
        사칙 연산을 계산하는 삼항 연산자 함수입니다.
        num1 : int
        num2 : int
        op : string 연산자 ( +, -, *, / )
    """
    return n1 + n2 if op == '+' else (n1 - n2 if op == '-' else (n1 * n2 if op == '*' else (n1 / n2 if op == '/' else 0)))
    
num1 = int(input('첫 번째 정수 입력 >> '))  
num2 = int(input('두 번째 정수 입력 >> '))  
op = (input('연산자 입력(+,-,*,/,%) >> '))
print('결과 : {}'.format(cal(num1, num2, op)))
print('결과 : {}'.format(calTernary(num1, num2, op)))
```

```python
# 함수 설명문
# doc string : 기능 설명을 작성하는 부분
# 정의된 함수 내에서 shift + tab 누르면 설명 확인 가능
# 함수 내부 첫 줄에 개행을 포함한 긴 줄 문자열 기호 ( """ content """) 로 작성한다

def greeting(name): # name : str, return : str
    """
        이름이 입력되면 인사를 하는 기능
        name (_str_): str 이름을 입력 받습니다.
        return (_str_)
    """
    return ""

greeting('지영')
```

```python
# 함수 return 키워드 정리
# 1. return 있는 함수의 반환값은 언제나 하나만 return 된다.

def add_sub(num1, num2):
    return num1 + num2, num1-num2

# packing :결과값을 합쳐줘 tuple 화된다.
a, b = add_sub(10, 7)
# unpacking : 따로 담을 수 있다.
print(a, b) # out : 17 3

# 2. 함수 안에서 return 을 만나면 함수가 종료된다.

def temp2():
    for i in range(3):
        return i
print(temp2()) # out : 0
```

```python
# 가변 매개 변수 (*)

def add(*num):
    total = 0
    for i in num:
        total += i
    print(total)        
    print(sum(num))
    print(num[0])
    print(num)
    return total
    
add(3,5,7,2,3,4,6,7,2,3,4,3,1,3,4,5,2,1,3,4,3)    
```

```python
# 함수 4가지 모형
# 1. 매개변수 x, 리턴 x
def hi():
    print('hi!')
# 2. 매개변수 o, 리턴 x
def hi(s):
    print(s)
# 3. 매개변수 x, 리턴 o
def hi():
    return print('hi!')
# 4. 매개변수 o, 리턴 o
def hi(s):
    return print(s)
```

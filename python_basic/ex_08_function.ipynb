```python
def divisor(n):   
    """   . """ 
    for i in range(1, int(n / 2 + 1)):
        if (n % i == 0):
            print(i, end=" ")
    print(n) 
            
divisor(25)            
```

```python
def cal(n1, n2, op):
    """ 
           .
        num1 : int
        num2 : int
        op : string  ( +, -, *, / )
    """
    result = 0
    if (op == '+'):
        return n1 + n2
    if (op == '-'):
        return n1 - n2
    if (op == '*'):
        return n1 * n2
    if (op == '/'):
        return n1 / n2
    if (op == '%'):
        return n1 % n2
    return 0
    
def calTernary(n1, n2, op):
    """ 
             .
        num1 : int
        num2 : int
        op : string  ( +, -, *, / )
    """
    return n1 + n2 if op == '+' else (n1 - n2 if op == '-' else (n1 * n2 if op == '*' else (n1 / n2 if op == '/' else 0)))
    
num1 = int(input('    >> '))  
num2 = int(input('    >> '))  
op = (input(' (+,-,*,/,%) >> '))
print(' : {}'.format(cal(num1, num2, op)))
print(' : {}'.format(calTernary(num1, num2, op)))
```

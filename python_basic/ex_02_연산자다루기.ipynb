### 산술 연산자

```python
# 나누기, 나머지, 나누기(몫) 구하기
# num1 변수 10 대입
# num2 변수 7 대입
num1 = 10
num2 = 7
print('나누기 결과 : ',num1 / num2)
print('나누기 몫 : ' , num1 // num2)
print('나머지 결과 : ', num1 % num2)
```

```python
# 문자열 더하기
str1 = '안녕'
str2 = '하세요'

print(str1 + str2)
```

```python
str3 = '10'
str4 = '7'
# 두 변수를 더하기
print(str3+str4)
```

```python
# num1 과 str4 는 더해질까?
# 숫자 + 문자열은 더해지지 않는다. 하지만 형변환을 하면 된다.
# 형변환 : int(str) , str(n)
# num1 = 10, str4 = '7'
print(str(num1) + str4)
print(num1 + int(str4))
```

### 복합 대입 연산자

```python
# = 대입 연산자
# += 더하고 대입
# -= 빼고 대입
# *= 곱하고 대입
# /= 나누고 대입
# %= 나머지 값 대입
# += 기준으로 설명

a = 3
b = 2

a += b # a = a + b

print(a)

# 파이썬 치환 : 새로운 임의의 변수 생성 없이 교환 가능하다
a,b = b, a
print(a, b)
```

### 비교 연산자

```python
# a > b : a 가 b 보다 크다
# a >= b : a 가 b 보다 크거나 같다
# a < b : a 는 b 보다 작다
# a <= b : a 는 b 보다 작거나 같다
# a == b : a 는 b 와 같다
# a != b : a 와 b 는 같지 않다
```

```python
# random module (library)
import random
a = random.randint(1, 10)
b = random.randint(1, 10)
print(a, b)
print(a > b)
print(a < b)
print(a == b)
print(a != b)
```

### 논리 연산자

```python
# and ( java : && ): 논리1 and 논리2
# a = 7, b = 3
# ex : a > b and a != b    => true, 1
# ex : a < b and a != b    => false, 0
#######################################
# or ( java : || ): 논리1 or 논리2
# a = 7, b = 3
# true or false   => true, 1
# true or true    => true, 1
#######################################
# not ( java : ! ): not 논리
# 논리를 뒤집는 것
```

```python
print(True and False) # out : False
print(2 > 1 and 2 == 1) # out : False
print(True or False) # out : True
print(2 == 2 or 2 < 1) # out : True
print(not True) # out : False
print(not 2 < 1) # out : True
```

### 멤버 연산자

```python
# 조건문에 자주 활용이 된다.
# in : 찾고자 하는 값(x)이 포함되어 있으면 True
# not in : 찾고자 하는 값(x)이 비포함되어 있으면 True

# x in 문자열
print('a' in 'aaaaabbbb') # out : True    

# x not in 문자열
print('d' not in 'bbbbbbbbcccc') # out : True
```

```python
abcd = ['a', 'b', 'c', 'd']
print('a' in abcd) # out : True
print('g' in abcd) # out : False
```

### 삼항 연산자

```python
# a if 조건식 else b
# True if (조건식) else False

score = 80

'합격' if (score > 60) else '불합격'
```

```python
a = int(input('정수입력 >> '))

b = '홀수' if (a % 2 == 1) else '짝수'

print(f'{a} 는(은) {b} 입니다.')
```

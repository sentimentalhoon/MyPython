#### 반복문 : 반복하여 수행하는 명령문

    - while, for


```python
number = 1

while number <= 3:
    print(number)
    number += 1
```

```python
# while 문 사용해서 "파이썬이 쉬워요." 문장을 13번 출력하기.

s = '파이썬이 쉬워요.'

number = 13
while number > 0:
    print(number, s)
    number -= 1
    
number = 13    
while True:
    print(number, s)
    number -= 1
    if (number <= 0):
        break
```

```python
while True:
    num1 = int(input("첫 번째 정수 입력 >> "))
    num2 = int(input("두 번째 정수 입력 >> "))
    plusNum = num1 + num2
    if (plusNum == 0):
        print('프로그램이 종료되었습니다.')
        break    
    print(f'두 정수의 합 : {plusNum}')
```

```python
import random
# (a: int, b: int) -> int
# Return random integer in range [a, b], including both end points.

ran = random.randint(1, 50)

while(True):
    inputNum = int(input("숫자를 입력하세요 >> "))
    if inputNum > ran :
        print("입력한 수보다 작은 수입니다.")
    elif inputNum < ran:
        print("입력한 수보다 큰 수 입니다.")
    else:
        print(f'{inputNum} 정답입니다.')
        break;    
```

```python
score_list = [ 90, 45, 70, 60, 55 ]


for i in score_list:
    if i >= 60:
        print(f' {score_list.index(i) + 1} 번 학생은 {i}점으로 합격입니다.')
    else:
        print(f' {score_list.index(i) + 1} 번 학생은 {i}점으로 불합격입니다.')
```

```python
# range(stop) -> range object range(start, stop[, step]) -> range object
# Return an object that produces a sequence of integers from start (inclusive) to stop 
# (exclusive) by step. range(i, j) produces i, i+1, i+2, ..., j-1. start defaults to 0, and stop is omitted! 
# range(4) produces 0, 1, 2, 3. These are exactly the valid indices for a list of 4 elements. 
# When step is given, it specifies the increment (or decrement).

for i in range(1, 100):
    print(i, end=" ")
    
# range 함수를 통해 index 번호 생성해서 사용하게 됨.    
```

```python
# range(start, end, step)
# range(0, n, 1)
# range(0, 10) : 0 부터 9 까지 1씩 증가.
# range(10) : 0 부터 9 까지 1씩 증가.
for i in range(len(score_list)):
    if score_list[i] >= 60:
        print(f' {i + 1} 번 학생은 {score_list[i]}점으로 합격입니다.')
    else:
        print(f' {i + 1} 번 학생은 {score_list[i]}점으로 불합격입니다.')
```

#### 이중 리스트 반복문 활용하기

```python
frd_list = [['강련', 28], ['건웅', 29], ['선웅', 26], ['주섭', 27], ['인우', 26], ['나행', 25]]
inputString = input('이름을 입력하시오 >> ')
for name, age in frd_list:
    if name == inputString:
        print(f'{name} 님의 나이는 {age} 세입니다.')
```

```python
# 빈 리스트 2개 선언
name_list = []
age_list = []

# 리스트 맨 뒤로 추가하는 함수 : targetList.append(value)

frd_list = [['강련', 28], ['건웅', 29], ['선웅', 26], ['주섭', 27], ['인우', 26], ['나행', 25]]

for name, age in frd_list:
    name_list.append(name)
    age_list.append(age)
    
# name_list, age_list print    
print(name_list)    
print(age_list)
```

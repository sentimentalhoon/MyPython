```python
list1 = [1, 50, 27, 19, 37, 33, 5, 3, 48]

oddNum = 0
evenNum = 0
for i in list1:
    if i % 2 == 1:
        oddNum += 1
    else:
        evenNum += 1

print(f'짝수 {evenNum} 개, 홀수 {oddNum} 개')
```

#### Question6

Traffic_Accident_2017.csv 데이터를 이용하여 ‘차대 사람’인 사건 중 야간에 발생한 사건에 대해 아래 주어진 조건을 지켜 다음과 같이 시각화 하시오.

※ 조건 내용 반드시 확인할 것  
 조건 1. 사고유형\_대분류가 ‘차대사람’ 이면서 발생시간대가 ‘야간’인 경우의 사고 건수 카운트  
 조건 2. y축의 범위는 0~200으로 설정할 것  
 조건 3. 차트 내부 글씨는 한글로 표시할 것


```python
font_path = 'C:\Windows\Fonts\malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)

traffic_accident = pd.read_csv('../data/csv/Traffic_Accident_2017.csv', encoding='euc-kr')
차대차_야간 = traffic_accident.query('사고유형_대분류 == "차대사람" and 주야 == "야간"')['요일'].value_counts()

차대차_야간 = 차대차_야간[['월', '화', '수', '목', '금', '토', '일']]

차대차_야간_x = 차대차_야간.index
차대차_야간_y = 차대차_야간

plt.title('요일별 야간 차대사람 교통사고 건수')
plt.bar(차대차_야간_x, 차대차_야간_y)
plt.xlabel('요일')
plt.ylabel('사고 건수')
plt.ylim(0, 200)
plt.show()
```

```python
economy = pd.read_csv("../data/csv/시도_성별_경제활동인구_총괄_KOSIS_2022.csv", encoding="euc-kr")
display(economy)
```

```python
# 고용률 24 non-null
# 실업률 28 non-null
display(economy.describe())
display(economy.info())
display(economy.size)
display(economy.head())

# 고용률, 실업률 컬럼 결측히가 있다면 해당 개수를 확인하는 방법
economy.isnull().sum()
# 고용률 : 생산가능연령(15세 이상) 중 특정 시점에 취업하고 있는 사람의 비율
# 실업률 : 경제 활동 참가자 중 실업 상태에 있는 사람의 비율
```

```python
# 고용률 결측치 처리해보기
# 특성공학 : 기존의 컬럼 활용하여 데이터 생성
# 고용률 : 취업자 / 15세이상인구
fifteen = economy['15세이상인구 (천명)']
취업자 = economy['취업자 (천명)']

고용률 = round((취업자 / fifteen) * 100, 1)

economy['고용률(%)'] = 고용률

# 실업률 -> 실업자/경제활동인구
실업자 = economy['실업자 (천명)']
경제활동인구 = economy['경제활동인구 (천명)']

실업률 = round((실업자/경제활동인구) * 100, 1)
economy['실업률(%)'] = 실업률

display(economy)
```

```python
# 시도별 고용률 평균 구해보기
round(economy['고용률(%)'].mean(), 2) # out : 63.44

# 시도별 고용률 평균이 가장 높은 지역 5개 확인해보기
economy.groupby(['행정구역(시도)']).agg(남녀고용률 = ('고용률(%)', 'mean')).sort_values(by = '남녀고용률', ascending=False).head()
emp_mean = economy[['행정구역(시도)','고용률(%)']].groupby('행정구역(시도)').mean()
emp_mean
# economy.sort_values(by='고용률(%)', ascending=False).head()
```

```python
economy_x = emp_mean.index
economy_y = emp_mean['고용률(%)']
plt.figure(figsize=(10, 3))
plt.bar(economy_x, economy_y, label='고용률 (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.ylim(55, 80)
plt.xlabel('지역명')
plt.ylabel('고용률 (%)')
plt.title('전국 평균 고용률 현황')
plt.show()
```

```python
plt.figure(figsize = (8, 4))
sns.barplot(x = economy_x, y = economy_y, label='고용률 (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.ylim(55, 80)
plt.title('전국 평균 고용률 현황')
plt.show()

display(economy)
```

```python
#economy_gender = economy[economy['성별'] == '남자'].groupby(['행정구역(시도)']).agg(실업률 = ('실업률(%)', 'sum'))
#economy.query('성별 == "남자"').groupby(['행정구역(시도)']).agg(실업률 = ('실업률(%)', 'mean')).sort_values('실업률', ascending=False)
economy_gender = economy.loc[economy['성별'] == '남자', ['행정구역(시도)','실업률(%)']]
gender_x = economy_gender['행정구역(시도)']
gender_y = economy_gender['실업률(%)']

sns.barplot(x = gender_x, y = gender_y, label='남성 실업률 (%)')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.title('전국 남성 실업률 현황')
plt.show()
```

#### Scatter plot(산점도), Pie Plot

```python
# 산점도 그래프 그려보기
# 산점도 점 = 버블 (버플사잊, 버블 색상 설정)
x = np.arange(1,4)
np.random.seed(2)
y = np.random.rand(3)

plt.scatter(x, y, s = 1, color='k')
plt.show()
```

```python
# 산점도 색상 bar 설정
plt.scatter(x, y, s = 5, c = range(3), cmap=plt.cm.Reds)
# cmap : matplotlib 색상 파레트 설정값
plt.colorbar()
plt.show()
```

```python
# Pie Plot
for i in y:
    print(f'{(i/y.sum()) * 100:.2f}')
    
plt.pie(y, labels=x, autopct='%.2f %%')
plt.show()
```

#### 실습 예제 문제

```python
### 1) 데이터 불러오기
traffic_accident = pd.read_csv('../data/csv/Traffic_Accident_2017.csv', encoding='euc-kr')
### 2) 데이터 정보 확인하기
#display(traffic_accident)
```

```python
### 3) 요일별 교통사고 시각화
day_traffic_accident = traffic_accident.groupby(['요일']).agg(day_accident_number = ('요일', 'count'))
day_traffic_accident_x = day_traffic_accident.index
day_traffic_accident_y = day_traffic_accident['day_accident_number']
plt.plot(day_traffic_accident_x, day_traffic_accident_y)
plt.show()
```

```python
### 4. 차대차 사건 중 사상자가 많이 발생한 시도를 알아보고 시각화
traffic_accident_by_sido = traffic_accident.query('사고유형_대분류 == "차대차"').groupby(['발생지시도']).agg(차대차 = ('발생년', 'count'))
traffic_accident_by_sido_x = traffic_accident_by_sido.index
traffic_accident_by_sido_y = traffic_accident_by_sido['차대차']
plt.bar(traffic_accident_by_sido_x, traffic_accident_by_sido_y)
plt.show()
```

```python
### 5) 교통사고가 가장 많이 발생하는 시간대를 알아보고 시각화 하라.

traffic_accident['시간'] = traffic_accident['발생년월일시'].astype(str).str[-2:]
# traffic_accident['시간'] = traffic_accident['발생년월일시'] % 100
time_sum = traffic_accident.groupby('시간').agg(Time = ('시간', 'count')).sort_values('시간', ascending=True)

time_sum_x = time_sum.index
time_sum_y = time_sum['Time']

plt.bar(time_sum_x, time_sum_y)
plt.show()

display(traffic_accident)
```

```python
### 6) 광주지역 법규 위반 사항별 사고 건수를 파이 그래프로 시각화 해보자
광주법규위반 = traffic_accident.query('발생지시도 == "광주"').groupby('법규위반').agg(위반수 = ('법규위반', 'count'))
광주법규위반_x = 광주법규위반.index
광주법규위반_y = 광주법규위반['위반수']

plt.pie(광주법규위반_y, labels=광주법규위반_x, autopct='%.2f %%')
plt.show()
```

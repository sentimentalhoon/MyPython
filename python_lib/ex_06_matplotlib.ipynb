```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Line Plot 선그래프

```python
# 배열 생성하여 데이터 준비
x = np.arange(1,6)
display(x)
np.random.seed(3)
y = np.random.randint(1, 10, size=5)
display(y)
```

```python
# 시각화
plt.plot(x, y, marker='p', color='#0accff', linestyle=':', linewidth='1', mec='r', mfc='k')
plt.show()
```

#### show() 를 이용하기

```python
# 여러 그래프를 한창에 그리기
# show()  : 도화지 (그래프창) 을 구분하는 역할
y2 = np.random.randint(1,6, size= 5)
y3 = np.random.randint(1,6, size= 5)

plt.plot(x,y)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```

- 그래프창 크기 조절하기

```python
plt.figure(figsize=(5, 2))
plt.plot(x,y)
plt.show()
```

- 한 창에 칸 나눠서 그래프 여러개 표시하기 : subplot()

```python
plt.figure(figsize=(15, 3))
plt.subplot(1, 3, 1)
plt.title('<encoding>') # 그래프 제목 달기
plt.xlabel('x label')
plt.ylabel('y label')
plt.plot(x,y)
plt.subplot(1, 3, 2)
plt.title('<2>') # 그래프 제목 달기
plt.xlabel('x')
plt.plot(x,y2)
plt.subplot(1, 3, 3)
plt.title('<3>') # 그래프 제목 달기
plt.xlabel('x')
plt.plot(x,y3)
plt.show()
```

- 예제1) 국가통계포털의 장래_인구변동_데이터를 활용하여 lineplot 으로 출생아수, 사망자수 시각화 해보기

```python
# 1. 데이터 불러오기
data = pd.read_csv("../data/csv/장래_인구변동_KOSIS.csv", encoding="euc-kr", index_col = '인구변동요인별')
```

```python
x = data.columns

# 출생아 수 (천명) : 데이터에 접근해보기
y1 = data.iloc[1]
# 사망자 수 (천명) : 데이터에 접근해보기
y2 = data.iloc[2]

plt.figure(figsize = (4,3))
plt.plot(y1.index, y1, marker='.')
plt.plot(y2.index, y2, marker='*')
plt.show()
```

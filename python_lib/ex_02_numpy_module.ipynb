### Numpy 배열 다루기

```python
# 1차원 array (배열) 생성하기

list1 = [1, 2, 3, 4, 5]

# np.array(data)

arr1 = np.array(list1)

# 2차원 array 생성하기

list2 = [[1,2,3], [4,5,6]]

# np.array(data)

arr2 = np.array(list2)

# 3차원 array 생성하기

list3 = [[[1,2,3], [4,5,6]] , [[1,2,3], [4,5,6]]]

# np.array(data)

arr3 = np.array(list3)
```

```python
# 배열 속성 확인하는 키워드
# 1차원 확인 ( 배열크기, 차원 수, 데이터 타입, 전체 요소 개수 )

print('배열 크기 : ', arr1.shape)
print('차원 수 : ', arr1.ndim)
print('데이터 타입 : ', arr1.dtype)
print('전체 요소 개수 : ', arr1.size)
```

```python
# 3차원 확인 ( 배열크기, 차원 수, 데이터 타입, 전체 요소 개수 )
print('배열 크기 : ', arr3.shape)
print('차원 수 : ', arr3.ndim)
print('데이터 타입 : ', arr3.dtype)
print('전체 요소 개수 : ', arr3.size)
```

### Numpy Data 접근 및 처리

```python
# 1 차원 배열 - 요소 1개 접근하기 (indexing)
array4 = np.array([10, 20, 30, 40, 50])
# 40 요소 데이터에 접근하기
array4[3]
# 10, 30 데이터 접근
array4[[0,2]] # array([10, 30])
```

```python
# 1차원 배열 - 여러 요소 접근하기 (slicing)

# 20~50 데이터 접근
array4[1:] # array([20, 30, 40, 50])

# 10, 30 데이터 접근
array4[:3:2] # array([10, 30])
```

```python
# 2차원 배열 접근
array5 = np.array([[10, 20, 30],[40, 50, 60]])
#array5 = array4.reshape((2,3))
array5
```

```python
# 2차원 배열 인덱싱
# 행, 열
array5[0][2]
array5[0,2]
```

```python
# 2차원 배열 슬라이싱
# 40, 50
array5[1][:2]
array5[1,:2]
array5[1, [0,1]]
```

```python
# 다양한 방법으로 array5 의 10, 30, 40, 60 의 데이터 접근하기
r1 = array5[:, ::2]
print(r1)  # out [[10 30] [40 60]]
r2 = array5[:, [0, 2]]
print(r2)  # out [[10 30] [40 60]]
r3 = array5[[[0], [1]], [[0, 2]]]
print(r3)  # out [[10 30] [40 60]]
```
